```python
peliculas = pd.read_csv(r"C:\Users\Nicolas\Downloads\Movie dataset\Movie dataset\movies.csv",
                        usecols=['genres','release_date','original_title','id'])
peliculas.info()
```

```python
peliculas.head()
```

```python
peliculas.dropna(inplace=True)
```

```python
peliculas["release_date"] = peliculas["release_date"].map(lambda x:pd.to_numeric(x.split('-',1)[0],downcast="integer"))
peliculas.info()
```

```python
peliculas["release_date"] = np.floor(peliculas.release_date/10).astype("int16")*10
```

```python
generosPorDecada = peliculas.loc[:,['genres','release_date']]
generosPorDecada.info()
```

```python
generosPorDecada["genres"].unique().size
```

```python
generosPorDecada["genres"] = generosPorDecada["genres"].str.split(',')
```

```python
generosPorDecada.head()
```

```python
generosPorDecada.head().explode("genres")
```

```python
generosPD = generosPorDecada.explode("genres").reset_index().pivot_table(index="release_date",columns="genres",aggfunc="count")
```

```python
generosPD = generosPD.fillna(0)
generosPD.info()
```

```python
generosPD.columns = generosPD.columns.get_level_values(1)
generosPD
```

```python
generosPD.idxmax(axis=1)
```
