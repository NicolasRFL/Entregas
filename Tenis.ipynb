```python
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
jugadores = pd.read_csv(r"C:\Users\Nicolas\Downloads\tenis\Player.csv")
jugadores.info()
```

```python
estadisticas = pd.read_csv(r"C:\Users\Nicolas\Downloads\tenis\Stats.csv")
estadisticas.info()
```

```python
partidos = pd.read_csv(r"C:\Users\Nicolas\Downloads\tenis\Match.csv")
partidos.info()
```

```python
partidos.set_index('match_id',inplace=True)
```

```python
jugadores = jugadores[jugadores['country']=='ARG']
```

```python
jugadores.set_index('player_id',inplace=True)
estadisticas.set_index('player_id',inplace=True)
```

```python
jugadoresArg = jugadores.join(estadisticas,how='inner',on='player_id')
```

```python
jugadoresArg.reset_index().set_index('match_id',inplace=True)
jugadoresArg = jugadoresArg.join(partidos,how='inner',on='match_id')
```

```python
jugadoresArg
```

```python
jugadoresTop10 = jugadoresArg.groupby(['name']).agg({'rank':'min'})
```

```python
jugadoresTop10 = jugadoresTop10.sort_values(by='rank').head(10)
```

```python
jugadoresTop10.rename(columns={'rank':'Mejor ranking'},inplace=True)
```

```python
jugadoresTop10
```

```python
jugadoresTop10Arg = jugadoresArg.set_index('name').join(jugadoresTop10,how='inner',on='name')
```

```python
jugadoresTop10Arg.groupby(['name','tournament']).agg({'match_minutes':'mean'})
```

```python
torneosATPPorJugador = jugadoresTop10Arg.groupby(['name','winner','tournament']).agg({'winner':'count'}).rename(columns={'winner':'Partidos ganados'}).reset_index()
```

```python
torneosATPPorJugador
```

```python
partidosATP = jugadoresTop10Arg.groupby(['name','winner','tournament']).agg({'winner':'count'}).rename(columns={'winner':'Partidos ganados'}).reset_index()
```

```python
totalPartidosPorJugador = partidosATP.groupby('name').agg({'Partidos ganados':'sum'}).reset_index()
totalPartidosPorJugador.rename(columns={'name':'Nombre'},inplace=True)
```

```python
lista=[]
def borrarRepetidos(x):
    if x in lista:
        return None
    lista.append(x)
    return x
partidosATP = partidosATP[partidosATP['winner']==True].sort_values(by='Partidos ganados',ascending=False).drop(columns='winner')
partidosATP['name'] = partidosATP['name'].map(borrarRepetidos)
```

```python
partidosATP.dropna(inplace=True)
partidosATP = partidosATP.set_index('name').join(jugadoresTop10).reset_index()
```

```python
partidosATP.rename(columns={'name':'Nombre','tournament':'Torneo con mas partidos ganados','Partido ganados':'Partidos ganados en su mejor torneo'},inplace=True)
```
