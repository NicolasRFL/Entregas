```python
crew = pd.read_csv(r"C:\Users\Nicolas\Downloads\Movie dataset\Movie dataset\crew.csv",usecols=['job','name','id','crew_id'])
crew.info()
```

```python
crew = crew.groupby(['name','id']).agg({'job':'unique'})
crew.info()
```

```python
crew.head()
```

```python
"""Al hacer inner se descarta informacion, pero como busco una sola persona lo considero la mejor opcion."""
castNCrew = cast.join(crew,on=['name','id'],how='inner',lsuffix='_left', rsuffix='_right')
castNCrew.head()
```

```python
castNCrew['characterCount'] = castNCrew.character.map(len)
```

```python
castNCrew['jobCount']=castNCrew.job.map(len)
```

```python
castNCrew
```

```python
castNCrew['nroles'] = castNCrew['characterCount']+castNCrew['jobCount']
castNCrew
```

```python
castNCrew.reset_index(level='id')
```
