```python
movies = pd.read_csv(r"C:\Users\Nicolas\Downloads\Movie dataset\Movie dataset\movies.csv",usecols=['production_companies','genres'])
movies
```

```python
movies.dropna(inplace=True)
movies.reset_index(inplace=True)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)
count_matrix =cv.fit_transform(movies.genres)
count_array = count_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = cv.get_feature_names_out())
```

```python
df = df[df.columns.difference(['foreign'])]
```

```python
movies.production_companies = movies.production_companies.str.split(',')
```

```python
df = df.div(df.sum(axis=1), axis=0)
```

```python
movies
```

```python
movies= movies.explode('production_companies')
```

```python
movies.reset_index(inplace=True)
```

```python
movies['production_companies'] = pd.to_numeric(movies['production_companies'])
```

```python
movies = pd.concat([df,movies['production_companies']],axis=1)
```

```python
movies = movies.groupby('production_companies').agg('sum')
```

```python
movies = movies[movies.sum(axis=1)>100]
```

```python
movies = movies.div(movies.sum(axis=1),axis=0)
```

```python
movies
```

```python
def entrop(x):
    if x==0:
        return 0
    return -x*np.log2(x)
movies = movies.applymap(entrop)
```

```python
movies['entropiaShannon'] = movies.sum(axis=1)
```

```python
companies = companies.rename({'id':'production_companies'},axis=1)
movies = movies.join(companies.set_index('production_companies'),on='production_companies')
```
